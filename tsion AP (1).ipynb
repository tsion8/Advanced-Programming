**1:**  **Class Design and Encapsulation**


```python
class Student:
    # Constructor
    def __init__(self, student_id, name, program, gpa):
        self.student_id = student_id
        self.name = name
        self.program = program
        self.__gpa = 0.0
        self.set_gpa(gpa)

    # Getter method
    def get_gpa(self):
        return self.__gpa

    # Setter method
    def set_gpa(self, gpa):
        if 0.0 <= gpa <= 4.0:
            self.__gpa = gpa
        else:
            print("Invalid GPA")

    # Display information
    def display_info(self):
        print("Student ID:", self.student_id)
        print("Name:", self.name)
        print("Program:", self.program)
        print("GPA:", self.__gpa)


# Test Cases
student1 = Student("UGR/001", "Tsion", "Computer Science", 3.8)
student1.display_info()

student1.set_gpa(3.9)
print("Updated GPA:", student1.get_gpa())
```

***REPORT***

Your code defines a Student class that models a university student with attributes for ID, name, program, and GPA. The constructor (__init__) initializes these attributes and ensures that the GPA is validated immediately by calling the setter method. The GPA attribute is made private (__gpa) to enforce encapsulation, meaning it cannot be accessed directly from outside the class. Instead, you provide a getter method (get_gpa) to safely retrieve the GPA and a setter method (set_gpa) to update it. The setter includes validation logic to ensure the GPA remains within the acceptable range of 0.0 to 4.0, printing an error message if the input is invalid. The display_info method prints all student details in a clear format, making it easy to view the stored data.


**2: Inheritance with Animal Classes**

```python

class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self):
        print("Animal sound")


class Dog(Animal):
    def speak(self):
        print(self.name, "says Woof")


class Cow(Animal):
    def speak(self):
        print(self.name, "says Moo")


class Bird(Animal):
    def speak(self):
        print(self.name, "says Tweet")


# Test Cases
animals = [
    Dog("Buddy", 3),
    Cow("Bella", 5),
    Bird("Coco", 1)
]

for animal in animals:
    animal.speak()
```

***Report***

Inheritance is implemented using the Animal parent class. Child classes override the speak() method, demonstrating polymorphism.so The program begins with a parent class called Animal, which defines the shared attributes name and age and a general method speak() that prints a generic message. Three child classes — Dog, Cow, and Bird — inherit from the Animal class and override the speak() method to provide animal-specific sounds. For example, the Dog class prints "Woof", the Cow class prints "Moo", and the Bird class prints "Tweet". In the test section, a list of different animal objects is created, and a loop iterates through them, calling the speak() method on each. Although the same method name is used, the output differs depending on the object type, which illustrates polymorphism — the ability of different objects to respond uniquely to the same method call. This design highlights how inheritance allows code reuse, method overriding customizes behavior, and polymorphism makes programs flexible and extensible.

**3: Bioinformatics Sequence System**

```python
# Parent class
class Sequence:

    # Constructor
    def __init__(self, seq_id, sequence):
        self.seq_id = seq_id
        self.sequence = sequence.upper()

    # Method to calculate sequence length
    def get_length(self):
        return len(self.sequence)

    # Method to display information
    def display_info(self):
        print("Sequence ID:", self.seq_id)
        print("Sequence:", self.sequence)
        print("Length:", self.get_length())


# Child class for DNA
class DNASequence(Sequence):

    # Validate DNA sequence
    def validate(self):
        allowed = "ATGC"
        for base in self.sequence:
            if base not in allowed:
                return False
        return True

    # Calculate GC content
    def gc_content(self):
        g = self.sequence.count("G")
        c = self.sequence.count("C")
        gc = ((g + c) / len(self.sequence)) * 100
        return gc


# Child class for RNA
class RNASequence(Sequence):

    # Validate RNA sequence
    def validate(self):
        allowed = "AUGC"
        for base in self.sequence:
            if base not in allowed:
                return False
        return True

    # Convert RNA to DNA
    def to_dna(self):
        return self.sequence.replace("U", "T")


# Child class for Protein
class ProteinSequence(Sequence):

    # Validate protein sequence
    def validate(self):
        allowed = "ACDEFGHIKLMNPQRSTVWY"
        for aa in self.sequence:
            if aa not in allowed:
                return False
        return True

    # Count amino acids
    def amino_acid_count(self):
        counts = {}

        for aa in self.sequence:
            if aa in counts:
                counts[aa] += 1
            else:
                counts[aa] = 1

        return counts



# Creating objects


dna = DNASequence("DNA001", "ATGCGCTA")
rna = RNASequence("RNA001", "AUGCGCUA")
protein = ProteinSequence("PRO001", "MKTAFLL")


# DNA Operations

print("DNA Sequence")
dna.display_info()
print("Valid DNA:", dna.validate())
print("GC Content:", dna.gc_content(), "%")

print("\n------------------")


# RNA Operations

print("RNA Sequence")
rna.display_info()
print("Valid RNA:", rna.validate())
print("Converted DNA:", rna.to_dna())

print("\n------------------")


# Protein Operations

print("Protein Sequence")
protein.display_info()
print("Valid Protein:", protein.validate())
print("Amino Acid Count:", protein.amino_acid_count())
```

***Report***

The Sequence class acts as a base class for biological sequences. DNA, RNA, and Protein classes inherit common properties while adding specialized biological methods.This program models biological sequences using object-oriented programming. It begins with a Sequence base class, which stores a sequence ID and the sequence string in uppercase, and provides a method to calculate the sequence length. Three specialized child classes inherit from it: DNASequence, RNASequence, and ProteinSequence. Each child class defines its own validate() method to check that the sequence contains only valid characters for its type. The DNA class adds a method to calculate GC content, which measures the proportion of guanine and cytosine bases in the sequence. 

**4: Library Management System**

```python
# Book class
class Book:
    def __init__(self, book_id, title, author):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.available = True   # Book is available by default

    def display_info(self):
        print(f"Book ID: {self.book_id}")
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Available: {self.available}")


# Member class
class Member:
    def __init__(self, member_id, name):
        self.member_id = member_id
        self.name = name
        self.borrowed_books = []   # List to store borrowed books

    def display_info(self):
        print(f"Member ID: {self.member_id}")
        print(f"Name: {self.name}")

        print("Borrowed Books:")
        if len(self.borrowed_books) == 0:
            print("No books borrowed")
        else:
            for book in self.borrowed_books:
                print("-", book.title)


# Library class
class Library:
    def __init__(self, name):
        self.name = name
        self.books = []      # List of Book objects
        self.members = []    # List of Member objects

    # Add a book
    def add_book(self, book):
        self.books.append(book)
        print(f"Book '{book.title}' added to library")

    # Register a member
    def register_member(self, member):
        self.members.append(member)
        print(f"Member '{member.name}' registered")

    # Borrow a book
    def borrow_book(self, member, book):
        if book.available:
            book.available = False
            member.borrowed_books.append(book)
            print(f"{member.name} borrowed '{book.title}'")
        else:
            print(f"Sorry, '{book.title}' is not available")

    # Return a book
    def return_book(self, member, book):
        if book in member.borrowed_books:
            book.available = True
            member.borrowed_books.remove(book)
            print(f"{member.name} returned '{book.title}'")
        else:
            print(f"{member.name} did not borrow '{book.title}'")


# -----------------------------------
# Create library
library = Library("University Library")

# Create books
book1 = Book(101, "Python Basics", "John Smith")
book2 = Book(102, "Data Structures", "Alice Brown")

# Add books to library
library.add_book(book1)
library.add_book(book2)

print("\n-------------------")

# Create members
member1 = Member(1, "Tsion")
member2 = Member(2, "Abel")

# Register members
library.register_member(member1)
library.register_member(member2)

print("\n-------------------")

# Borrow books
library.borrow_book(member1, book1)

# Try borrowing unavailable book
library.borrow_book(member2, book1)

print("\n-------------------")

# Display member info
member1.display_info()

print("\n-------------------")

# Return book
library.return_book(member1, book1)

# Borrow again after return
library.borrow_book(member2, book1)
```

***Report***

The Library class demonstrates composition by storing Book and Member objects. State management is implemented through book availability tracking.The Book class stores information about each book, including its ID, title, author, and whether it is available. By default, every book starts as available. The Member class represents a library member with an ID, name, and a list of borrowed books. It does not directly manage borrowing or returning but keeps track of which books the member currently has.
The Library class acts as the central manager. It maintains collections of books and members and provides methods to add books, register members, borrow books, and return books. When a member borrows a book, the system checks if the book is available. If it is, the book’s availability is set to False and the book is added to the member’s borrowed list. If the book is unavailable, the system prevents borrowing and prints a message. Returning a book reverses this process: the book is removed from the member’s borrowed list and marked as available again

**5: Abstract Base Class and Polymorphism**

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    """Abstract base class for geometric shapes."""

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)


class Triangle(Shape):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def area(self):
        # Heron's formula
        s = (self.a + self.b + self.c) / 2
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))

    def perimeter(self):
        return self.a + self.b + self.c


# --- Test Cases ---
shapes = [
    Circle(5),
    Rectangle(4, 6),
    Triangle(3, 4, 5)
]

for shape in shapes:
    print(f"{shape.__class__.__name__} -> Area: {shape.area():.2f}, Perimeter: {shape.perimeter():.2f}")
```

***Report***

The Shape class is abstract and defines a common interface. Child classes implement required methods, demonstrating abstraction and polymorphism.This program defines an abstract base class called Shape, which establishes a common interface for geometric objects. The class declares two abstract methods, area() and perimeter(), ensuring that any child class must implement them. Because Shape is abstract, it cannot be instantiated directly; instead, it serves as a blueprint for specific shapes.

Two concrete classes — Circle and Rectangle — inherit from Shape and provide their own implementations of the abstract methods. The circle calculates its area using the formula πr² and its perimeter using 2πr, while the rectangle computes its area as length × width and its perimeter as 2 × (length + width). These implementations demonstrate how different shapes can share the same interface but produce unique results.

**6: Variant Annotation System**

```python
class Variant:
    """Base class for genetic variants."""

    def __init__(self, chromosome, position, ref, alt):
        self.chromosome = chromosome
        self.position = position
        self.ref = ref
        self.alt = alt

    def classify(self):
        return "Generic Variant"

    def length(self):
        return abs(len(self.alt) - len(self.ref))

    def summary(self):
        return (f"Chromosome: {self.chromosome}, Position: {self.position}, "
                f"Ref: {self.ref}, Alt: {self.alt}, "
                f"Type: {self.classify()}, Length: {self.length()}")


class SNP(Variant):
    def classify(self):
        return "SNP"

    def length(self):
        return 1  # SNP always changes a single base


class Insertion(Variant):
    def classify(self):
        return "Insertion"

    def length(self):
        return len(self.alt) - len(self.ref)


class Deletion(Variant):
    def classify(self):
        return "Deletion"

    def length(self):
        return len(self.ref) - len(self.alt)


# --- Test Cases ---
variants = [
    SNP("chr1", 12345, "A", "G"),
    Insertion("chr2", 67890, "A", "ATCG"),
    Deletion("chr3", 13579, "ATCG", "A")
]

for v in variants:
    print(v.summary())
```

***Report***

This program models a simplified Variant Annotation System using object-oriented programming. The base class Variant stores common attributes such as chromosome, position, reference allele, and alternate allele. It also provides generic methods to classify the variant and compute its length.

Three child classes — SNP, Insertion, and Deletion — inherit from Variant and override the classify() method to identify the correct variant type. Each subclass also customizes the length() method to reflect biological rules: SNPs always have length 1, insertions calculate the difference between alternate and reference alleles, and deletions calculate the difference between reference and alternate alleles.

**7: Alternative Constructors**

```python
class Genome:

    def __init__(self, name, species, size):
        self.name = name
        self.species = species
        self.size = int(size)

    @classmethod
    def from_csv_line(cls, line):
        name, species, size = line.split(",")
        return cls(name, species, size)

    @classmethod
    def from_dict(cls, data):
        return cls(data["name"], data["species"], data["size"])


# Test Cases

g1 = Genome.from_csv_line("NDama,Bos taurus,2800")
print(g1.name, g1.species, g1.size)

data = {
    "name": "Human",
    "species": "Homo sapiens",
    "size": 3200
}

g2 = Genome.from_dict(data)
print(g2.name, g2.species, g2.size)
```

***Report***

Class methods provide alternative constructors for structured input such as CSV lines and dictionaries.The test cases demonstrate both approaches: one genome is created from a CSV line ("NDama,Bos taurus,2800") and another from a dictionary ({"name": "Human", "species": "Homo sapiens", "size": 3200}). Both objects store the correct attributes and can be accessed consistently.

**8: Static Methods vs Class Methods**

```python
class DNAUtils:

    valid_bases = "ATGC"

    @staticmethod
    def is_valid_dna(sequence):
        sequence = sequence.upper()
        return all(base in DNAUtils.valid_bases for base in sequence)

    @staticmethod
    def gc_content(sequence):
        sequence = sequence.upper()
        gc = sequence.count("G") + sequence.count("C")
        return (gc / len(sequence)) * 100

    @classmethod
    def update_valid_bases(cls, new_bases):
        cls.valid_bases = new_bases


# Test Cases
print(DNAUtils.is_valid_dna("ATGC"))
print(DNAUtils.gc_content("ATGC"))

DNAUtils.update_valid_bases("ATGCU")
print(DNAUtils.valid_bases)
```

***Report***

Static methods provide utility operations that do not require object creation. The class variable stores shared biological rules.his program defines a DNAUtils class to illustrate the difference between static methods and class methods. The static methods is_valid_dna() and gc_content() perform sequence validation and GC content calculation independently of class state, making them reusable utility functions. The class variable valid_bases stores the set of acceptable bases, and the class method update_valid_bases() allows updating this shared configuration. Test cases demonstrate validation of sequences, calculation of GC content, and updating valid bases to include additional symbols like "U".

***9: Operator Overloading***

```python
class Sequence:

    def __init__(self, seq_id, sequence):
        self.seq_id = seq_id
        self.sequence = sequence.upper()

    def __len__(self):
        return len(self.sequence)

    def __str__(self):
        return f"{self.seq_id}: {self.sequence}"

    def __eq__(self, other):
        return self.sequence == other.sequence

    def __add__(self, other):
        return Sequence(
            self.seq_id + "_" + other.seq_id,
            self.sequence + other.sequence
        )


# Test Cases
seq1 = Sequence("SEQ1", "ATGC")
seq2 = Sequence("SEQ2", "GGTA")
seq3 = Sequence("SEQ3", "ATGC")

print(len(seq1))
print(seq1)
print(seq1 == seq3)

combined = seq1 + seq2
print(combined)
```

***Report***

Magic methods allow Sequence objects to behave like native Python objects. Operator overloading improves usability and readability.This program defines a Sequence class that demonstrates operator overloading using Python’s magic methods. The __len__() method returns the sequence length, while __str__() provides a human-readable description of the sequence. The __eq__() method compares two sequences for equality, and __add__() joins two sequences together, producing a new Sequence object.
