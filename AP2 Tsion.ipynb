**Part 3:** Loop Practice (10 marks)

```python
for number in range(1, 21):
    if number % 3 == 0 and number % 5 == 0:
        print("FizzBuzz")
    elif number % 3 == 0:
        print("Fizz")
    elif number % 5 == 0:
        print("Buzz")
    else:
        print(number)
```

**Simple Calculator**

**Create a program that:**

**1. Takes two numbers from the user**

 ***2. Asks for an operation (+, -, *, /)***

**3. Prints the result**

**1,Addition**

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

operation = input("Enter operation (+): ")

if operation == "+":
    result = num1 + num2
else:
    result = "Invalid operation"

print("Result:", result)
```

**2,Substraction**

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

operation = input("Enter operation (-): ")

if operation == "-":
    result = num1 - num2
else:
    result = "Invalid operation"

print("Result:", result)
```

**3,Mulltiplication**

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

operation = input("Enter operation (*): ")

if operation == "*":
    result = num1 * num2
else:
    result = "Invalid operation"

print("Result:", result)
```
