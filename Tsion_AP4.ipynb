# Why is GenBank More Informative than FASTA for Genome Annotation Analysis?

GenBank records are more informative than FASTA files because they contain not only the biological sequence, but also detailed annotations and metadata that describe the biological meaning of the sequence.

FASTA files mainly store raw nucleotide or protein sequences with minimal description, while GenBank files provide structured information useful for genome annotation and bioinformatics analysis.

| Feature | FASTA | GenBank |
|---|---|---|
| Sequence data | Yes | Yes |
| Gene annotations | No | Yes |
| CDS regions | No | Yes |
| Protein products | No | Yes |
| Organism metadata | Limited | Detailed |
| Scientific references | No | Yes |

---

## 1. Gene Names

FASTA files usually contain only a short header line:

```text
>seq1
ATGCGTACGTAG
gene            100..900
                /gene="cytb"
                2. CDS Coordinates

GenBank files contain CDS (Coding Sequence) coordinates that indicate where protein-coding regions start and end.
CDS             120..850
These coordinates help researchers:

extract coding regions
identify exons
analyze mutations
translate DNA into proteins

FASTA files do not contain positional information.
3. Protein Products

GenBank records often describe the protein encoded by the gene.

Example:

/product="cytochrome b"

This annotation explains the biological product generated from the sequence.

FASTA files only provide sequence information and do not explain what protein or function is associated with the sequence.
4. Organism Metadata

GenBank includes detailed metadata about the organism:

ORGANISM  Bos taurus

Additional metadata may include:

taxonomy
strain
tissue source
chromosome number
mitochondrial or nuclear origin

This information is essential for:

evolutionary studies
comparative genomics
phylogenetic analysis

FASTA headers may contain limited organism names but lack standardized metadata.

5. References and Publications

GenBank files include scientific references related to the sequence.

Example:

REFERENCE   1
AUTHORS     Smith et al.
TITLE       Complete mitochondrial genome analysis

These references allow researchers to:

verify experimental evidence
trace sequence origin
consult published studies

FASTA files do not contain publication or citation information.

Conclusion

GenBank is more informative than FASTA because it combines sequence data with rich biological annotations, genomic features, metadata, and literature references.

This makes GenBank highly useful for:

genome annotation
gene extraction
protein analysis
mutation studies
comparative genomics

In contrast, FASTA is mainly suitable for storing and sharing raw sequence data.

# ***Exercise 7 — Mutation Interpretation***

````python
from Bio import Align

# Define sequences
seq1 = "ATGCCCTAG"
seq2 = "ATGCGCTAG"

# Create aligner object
aligner = Align.PairwiseAligner()

# Perform alignment
alignments = aligner.align(seq1, seq2)

# Print alignment information
print("Number of alignments:", len(alignments))
print("Best alignment score:", alignments[0].score)

# Print best alignment
print("\nBest Alignment:")
print(alignments[0])

# ---------------------------------------------------
# Identify sequence differences
# ---------------------------------------------------

print("\nDifferences Found:")

for i in range(len(seq1)):
    if seq1[i] != seq2[i]:
        print(f"Position {i+1}: {seq1[i]} → {seq2[i]}")
````

# ***Exercise 10 — Improve the Pipeline***

````python
from Bio import SeqIO
import pandas as pd

# Function to calculate GC content
def calculate_gc(sequence):
    sequence = str(sequence).upper()

    gc_count = sequence.count("G") + sequence.count("C")

    return (gc_count / len(sequence)) * 100


# Function to calculate AT content
def calculate_at(sequence):
    sequence = str(sequence).upper()

    at_count = sequence.count("A") + sequence.count("T")

    return (at_count / len(sequence)) * 100


# Function to classify GC category
def gc_category(gc):

    if gc >= 60:
        return "High GC"

    elif gc >= 40:
        return "Medium GC"

    else:
        return "Low GC"


# Improved pipeline function
def analyze_fasta(input_fasta, output_csv):

    results = []

    for record in SeqIO.parse(input_fasta, "fasta"):

        seq = record.seq
        seq_string = str(seq).upper()

        # GC and AT content
        gc_content = calculate_gc(seq)
        at_content = calculate_at(seq)

        # Reverse complement
        reverse_comp = str(seq.reverse_complement())

        # Check for ambiguous bases
        contains_ambiguous = "N" in seq_string

        # Translation
        translated = ""
        stop_codons = 0

        if seq_string.startswith("ATG") and len(seq_string) % 3 == 0:

            translated = str(seq.translate(to_stop=False))

            # Count stop codons (*)
            stop_codons = translated.count("*")

        # Save results
        results.append({
            "id": record.id,
            "description": record.description,
            "length": len(seq),
            "gc_content": round(gc_content, 2),
            "at_content": round(at_content, 2),
            "gc_category": gc_category(gc_content),
            "starts_with_ATG": seq_string.startswith("ATG"),
            "length_multiple_of_3": len(seq_string) % 3 == 0,
            "contains_N": contains_ambiguous,
            "reverse_complement": reverse_comp,
            "translated_sequence": translated,
            "stop_codon_count": stop_codons
        })

    # Create DataFrame
    df = pd.DataFrame(results)

    # Save CSV
    df.to_csv(output_csv, index=False)

    return df


# Run pipeline
pipeline_results = analyze_fasta(
    "example_sequences.fasta",
    "sequence_summary.csv"
)

# Display results
print(pipeline_results)
````

# Mitochondrial Genome Comparison

## Objective

Retrieve mitochondrial genome sequences from multiple species using Biopython and compare:

- sequence length
- GC content

This project demonstrates:
- Entrez database searching
- FASTA retrieval
- sequence analysis
- summary table generation

---



````python
from Bio import Entrez
from Bio import SeqIO
import pandas as pd

# Set email
Entrez.email = "legessetsion8@gmail.com"

# Species list
species_list = [
    "Bos taurus",
    "Capra hircus",
    "Ovis aries",
    "Sus scrofa"
]

# GC content function
def calculate_gc(sequence):

    sequence = str(sequence).upper()

    gc_count = sequence.count("G") + sequence.count("C")

    return (gc_count / len(sequence)) * 100


# Store results
results = []

# Process each species
for species in species_list:

    print(f"Searching for {species} mitochondrial genome...")

    # Corrected search query
    handle = Entrez.esearch(
        db="nucleotide",
        term=f'"{species}"[Organism] AND mitochondrion[Title] AND complete genome',
        retmax=1
    )

    search_results = Entrez.read(handle)
    handle.close()

    id_list = search_results["IdList"]

    # Skip if no records found
    if len(id_list) == 0:
        print("No records found.")
        continue

    print("NCBI ID:", id_list[0])

    # Fetch FASTA sequence
    handle = Entrez.efetch(
        db="nucleotide",
        id=id_list[0],
        rettype="fasta",
        retmode="text"
    )

    record = SeqIO.read(handle, "fasta")
    handle.close()

    seq = record.seq

    # Calculate statistics
    length = len(seq)
    gc_content = calculate_gc(seq)

    a_count = seq.count("A")
    t_count = seq.count("T")
    g_count = seq.count("G")
    c_count = seq.count("C")

    # Save results
    results.append({
        "Species": species,
        "Sequence_Length": length,
        "GC_Content": round(gc_content, 2),
        "A_Count": a_count,
        "T_Count": t_count,
        "G_Count": g_count,
        "C_Count": c_count
    })

# Create dataframe
df = pd.DataFrame(results)

# Display results
print("\nMitochondrial Genome Summary")
print(df)

# Save CSV
df.to_csv("mitochondrial_genome_summary.csv", index=False)

print("\nSummary saved to mitochondrial_genome_summary.csv")
````
